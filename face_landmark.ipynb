```python
import cv2
import numpy as np
import dlib
import matplotlib.pyplot as plt

# function to convert dlib.full_object_detection to numpy array
def shape_to_np(shape, dtype="int"):
	coords = np.zeros((68, 2), dtype=dtype)
	for i in range(0, 68):
		coords[i] = (shape.part(i).x, shape.part(i).y)
	return coords

# reading an image and converting it to grayscale
image = cv2.imread("test/1.jpg")
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

# loading the classifiers with respected files
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
detector = dlib.get_frontal_face_detector()
faces = detector(gray)
# looping through each detected faces and drawing rectangle around the face and circles around the feature points
# for face in faces:
#     landmarks = predictor(gray, face)
#     points = shape_to_np(landmarks)
#     points_dict = {key:list(value) for key, value in enumerate(points)}
#     x1=landmarks.part(36).x
#     x2=landmarks.part(39).x
#     y1=landmarks.part(37).y
#     y2=landmarks.part(40).y
#     lefteye=image[y1-5:y2+5,x1-5:x2+5]

#     plt.imshow(lefteye)
#     plt.show()
```

```python
# Lips
for face in faces:
    landmarks = predictor(gray, face)
    points = shape_to_np(landmarks)
    # points_dict = {key:list(value) for key, value in enumerate(points)}
    x1=landmarks.part(48).x
    x2=landmarks.part(54).x
    y1=landmarks.part(51).y
    y2=landmarks.part(58).y
    print(x1,x2,y1,y2)
    crop=image[y1-10:y2+10,x1-10:x2+10]


    # outputNameofImage = "lips_1.jpg"
    # print("Saving output image to", outputNameofImage)
    # cv2.imwrite(outputNameofImage, crop)
    plt.imshow(cv2.cvtColor(crop,cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
# both eyes
for face in faces:
    landmarks = predictor(gray, face)
    points = shape_to_np(landmarks)
    # points_dict = {key:list(value) for key, value in enumerate(points)}
    x1=landmarks.part(37).x
    x2=landmarks.part(47).x
    y1=landmarks.part(36).y
    y2=landmarks.part(45).y
    print(x1,x2,y1,y2)
    crop=image[y1-20:y2+20,x1-20:x2+20]


    outputNameofImage = "eyes_1.jpg"
    print("Saving output image to", outputNameofImage)
    cv2.imwrite(outputNameofImage, crop)
    plt.imshow(cv2.cvtColor(crop,cv2.COLOR_BGR2RGB))
    plt.show()
```
